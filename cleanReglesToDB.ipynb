```python
import pandas as pd
```

```python
REGLES = './_output/regles.csv'
REGLES_CLEAN = './_output/regles_clean.csv'
```

```python
df_regles = pd.read_csv(REGLES)
df_regles.set_index('source_id', inplace=True)
df_regles.head()
```

```python
df_regles.drop('order_niveau_regle', axis=1, inplace=True)
```

```python
def falseToEmpty(val):
    if (val == 'False' or val == False):
        return ''
    elif(type(val) is str and  len(val) > 250): # si jamais la taille est supérieure à un string
        return val[:250]
    return val
```

```python
df_regles['exception'] = df_regles.apply(lambda row: falseToEmpty(row.exception), axis=1)
df_regles['cible_si_independant'] = df_regles.apply(lambda row: falseToEmpty(row.cible_si_independant), axis=1)
df_regles['min'] = df_regles.apply(lambda row: falseToEmpty(row['min']), axis=1)
df_regles['max'] = df_regles.apply(lambda row: falseToEmpty(row['max']), axis=1)
df_regles.head()
```

```python
def cleanTrueFalse(val):
    if (val == 'True' or val == True):
        return 1
    else:
        return 0
```

```python
df_regles['flag_chiffre_independant_des_benefs'] = df_regles.apply(lambda row: cleanTrueFalse(row.flag_chiffre_independant_des_benefs), axis=1)
df_regles['flag_optionel'] = df_regles.apply(lambda row: cleanTrueFalse(row.flag_optionel), axis=1)
df_regles['flag_cible_multiple'] = df_regles.apply(lambda row: cleanTrueFalse(row.flag_cible_multiple), axis=1)
df_regles['flag_delegation'] = df_regles.apply(lambda row: cleanTrueFalse(row.flag_delegation), axis=1)
df_regles.head()
```

```python
df_regles.to_csv(REGLES_CLEAN)
```
